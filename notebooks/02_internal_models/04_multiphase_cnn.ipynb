# W4 multiphase 3D CNN

**Scientific purpose.** Define the historical four-channel imaging branch: a randomly
initialized MONAI 3D ResNet-18 receiving independently lesion-centered P, C1, C2, and C3
crops with shape `4 x 96 x 96 x 96`.

**Runnability classification:** historical training workflow requiring private processed
arrays and private patient-level splits. Definitions are inspectable, but training is not
launched automatically.

**Inputs:** private processed arrays and patient-level folds. **Private outputs when run:**
fold checkpoints and OOF/test probabilities outside the repository.

The channels were not explicitly registered. Joint flips preserve the already-stacked
tensor transformation but do not establish anatomical correspondence between source
phases. W4 fold 0 from the historical run was not retained.


```python
from pathlib import Path
import sys


def locate_repository(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in (current, *current.parents):
        if (candidate / "pyproject.toml").is_file() and (candidate / "configs").is_dir():
            return candidate
    raise RuntimeError("Run this notebook from within the cloned repository.")


REPO_ROOT = locate_repository()
SRC_ROOT = REPO_ROOT / "src"
if str(SRC_ROOT) not in sys.path:
    sys.path.insert(0, str(SRC_ROOT))

from liver_tumor_pipeline.config import load_config, require_path

CONFIG_PATH = REPO_ROOT / "configs" / "paths.yaml"
if not CONFIG_PATH.is_file():
    raise FileNotFoundError(
        "Create an untracked configs/paths.yaml from configs/paths.example.yaml and "
        "set the documented environment variables before running this workflow."
    )

CONFIG = load_config(CONFIG_PATH)
PRIVATE_ARTIFACT_ROOT = require_path(CONFIG, "private_artifact_root", must_exist=False)
OUTPUT_ROOT = require_path(CONFIG, "output_root", must_exist=False)

import json
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from monai.networks.nets import resnet18
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

SPLIT_ROOT = PRIVATE_ARTIFACT_ROOT / "splits"
PROCESSED_ROOT = PRIVATE_ARTIFACT_ROOT / "processed" / "internal"
for required in (SPLIT_ROOT / "train_val_test.json", SPLIT_ROOT / "fold_0.json"):
    if not required.is_file():
        raise FileNotFoundError(
            "Private split definitions are unavailable; run notebook 01 in authorized storage."
        )
if not PROCESSED_ROOT.is_dir():
    raise FileNotFoundError(
        "Private processed arrays are unavailable; run notebook 02 in authorized storage."
    )

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = False
torch.backends.cudnn.benchmark = True
```

## Four-channel dataset and augmentation

Each channel is normalized independently over nonzero liver voxels. The same random
in-plane flip is applied to every channel; intensity jitter is sampled separately for
each channel. Validation and test samples are not augmented.


```python
class MultiphaseDataset(Dataset):
    def __init__(self, patient_keys, processed_root: Path, *, train: bool = False):
        self.patient_keys = list(patient_keys)
        self.processed_root = Path(processed_root)
        self.train = train

    def __len__(self):
        return len(self.patient_keys)

    @staticmethod
    def normalize_per_channel(volumes: np.ndarray) -> np.ndarray:
        output = np.zeros_like(volumes, dtype=np.float32)
        for channel in range(volumes.shape[0]):
            volume = volumes[channel]
            mask = volume > 0
            if mask.sum() < 100:
                output[channel] = volume.astype(np.float32)
            else:
                values = volume[mask]
                output[channel][mask] = (values - values.mean()) / (values.std() + 1e-6)
        return output

    @staticmethod
    def augment(volumes: np.ndarray) -> np.ndarray:
        if random.random() < 0.5:
            volumes = volumes[:, ::-1].copy()
        if random.random() < 0.5:
            volumes = volumes[:, :, ::-1].copy()
        if random.random() < 0.5:
            for channel in range(volumes.shape[0]):
                scale = 1.0 + (random.random() - 0.5) * 0.2
                shift = (random.random() - 0.5) * 0.2
                nonzero = volumes[channel] != 0
                volumes[channel][nonzero] = volumes[channel][nonzero] * scale + shift
        return volumes

    def __getitem__(self, index):
        patient_key = self.patient_keys[index]
        path = self.processed_root / f"{patient_key}.npz"
        if not path.is_file():
            raise FileNotFoundError("A required private processed patient array is unavailable.")
        with np.load(path, allow_pickle=False) as archive:
            volumes = archive["ct"].astype(np.float32)
            label = int(archive["label"])
        if volumes.shape != (4, 96, 96, 96):
            raise ValueError(f"Unexpected multiphase input shape: {volumes.shape}")
        volumes = self.normalize_per_channel(volumes)
        if self.train:
            volumes = self.augment(volumes)
        return torch.from_numpy(volumes.astype(np.float32)), label, patient_key


def make_w4_model() -> nn.Module:
    return resnet18(
        spatial_dims=3,
        n_input_channels=4,
        num_classes=3,
        feed_forward=True,
    )
```

```python
from sklearn.utils.class_weight import compute_class_weight

BATCH_SIZE = 4
NUM_EPOCHS = 40
LEARNING_RATE = 1e-4
WEIGHT_DECAY = 1e-5


def evaluate(model, loader, device):
    model.eval()
    labels, probabilities = [], []
    with torch.no_grad():
        for images, targets, _patient_keys in loader:
            logits = model(images.to(device))
            probabilities.append(torch.softmax(logits, dim=1).cpu().numpy())
            labels.append(targets.numpy())
    y_true = np.concatenate(labels)
    probs = np.concatenate(probabilities)
    predictions = probs.argmax(axis=1)
    return {
        "auc": roc_auc_score(y_true, probs, multi_class="ovr", average="macro"),
        "accuracy": accuracy_score(y_true, predictions),
        "macro_f1": f1_score(y_true, predictions, average="macro"),
        "labels": y_true,
        "probabilities": probs,
        "predictions": predictions,
    }


def train_one_fold(model, train_loader, validation_loader, development_labels, device):
    # Historical class weights were computed from all 222 development labels, not only the
    # fold-training subset. This is preserved and disclosed as a CV-isolation limitation.
    weights = compute_class_weight(
        class_weight="balanced",
        classes=np.array([0, 1, 2]),
        y=np.asarray(development_labels),
    )
    criterion = nn.CrossEntropyLoss(
        weight=torch.as_tensor(weights, dtype=torch.float32, device=device)
    )
    optimizer = torch.optim.AdamW(
        model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY
    )
    best_auc = -np.inf
    best_state = None
    best_metrics = None
    for _epoch in range(NUM_EPOCHS):
        model.train()
        for images, targets, _patient_keys in train_loader:
            optimizer.zero_grad(set_to_none=True)
            loss = criterion(model(images.to(device)), targets.to(device))
            loss.backward()
            optimizer.step()
        metrics = evaluate(model, validation_loader, device)
        if metrics["auc"] > best_auc:
            best_auc = metrics["auc"]
            best_metrics = metrics
            best_state = {
                key: value.detach().cpu().clone() for key, value in model.state_dict().items()
            }
    if best_state is None or best_metrics is None:
        raise RuntimeError("No finite validation checkpoint was selected.")
    return best_state, best_metrics
```

## Deliberate execution boundary

Call `train_one_fold` only in an authorized training environment after constructing
fold-specific loaders from private split files. The same CV checkpoint-selection and
development-wide class-weight limitations described for W3 apply here. The verified
aggregate W4 results are released in `results/aggregate/internal_performance.csv`.
Neither phase-drop analysis nor Grad-CAM should be interpreted as voxelwise temporal
tracking across registered acquisitions.
